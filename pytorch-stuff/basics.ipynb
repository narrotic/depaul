# ðŸ§© Key Idea: Using -1 in reshape

When reshaping, you donâ€™t need to specify every dimension manually. Why? Because PyTorch can figure out the missing one automatically using: total elements Ã· product of other dimensions. You tell PyTorch: â€œFigure this one out for me.â€ By writing -1.

### Suppose you have a tensor x with 12 elements.

x = torch.arange(12) # [0,1,2,...,11]

You want to reshape it to (3, 4):

- Normal way: \*
  x.reshape(3, 4)

- But this is equivalent:\*
  x.reshape(-1, 4)

Here:

PyTorch sees you want: some rows = ? columns = 4
total elements = 12
so number of rows = 12 / 4 = 3 _Automatically inferred._
So it becomes (3, 4).

Another equivalent:
x.reshape(3, -1)

This time:
rows = 3
columns = ?

So:
columns = 12 / 3 = 4
It becomes (3, 4) again.

### âœ”ï¸ Rule

You can use exactly one -1 in reshape.
Because only one dimension can be inferred.

### âœ¨ WHY THIS IS USEFUL

When coding, you often don't know one dimension easily.

Example:
You know each sample has 784 pixels (MNIST), but batch size varies.

So you can write:

x.reshape(-1, 784)

PyTorch will set the batch size automatically.

This makes code cleaner, safer, and dynamic.

### ðŸ§ª Bonus: Examples with real shapes

Say we have a tensor of shape (2, 3, 4)
Total elements: 2 _ 3 _ 4 = 24

We can reshape to:
x.reshape(4, 6)
x.reshape(6, 4)
x.reshape(8, 3)
x.reshape(3, 8)
x.reshape(-1, 6) # => (4, 6)

# NumPy Slicing (Complete Explanation in One Cell)

NumPy slicing lets you extract or modify parts of an array. The general form is:

`array[start:stop:step]`

- `start`: index to begin (inclusive), default = 0
- `stop`: index to end (exclusive), default = end of array
- `step`: how much to skip, default = 1

Examples:

- `x[:]` â†’ all elements
- `x[1:4]` â†’ elements at indices 1,2,3
- `x[::2]` â†’ every second element
- `x[::-1]` â†’ reversed array

---

## Multi-Dimensional Slicing

For 2D arrays, each dimension has its own slice:

`X[row_slice, col_slice]`

Examples:

- `X[:, :]` â†’ all rows, all columns
- `X[1, :]` â†’ row 1, all columns
- `X[:, 2]` â†’ column 2, all rows
- `X[1:3, 0:2]` â†’ rows 1â€“2, cols 0â€“1

---

## Using `:` in Different Ways

### 1. Select first N rows:

`X[:2, :]`

- rows 0 and 1
- all columns

### 2. Select first M columns:

`X[:, :3]`

- all rows
- columns 0,1,2

### 3. Select with step:

`X[::2, :]`

- every second row

### 4. Reverse rows/columns:

- rows reversed: `X[::-1, :]`
- columns reversed: `X[:, ::-1]`

---

## Negative Indices

Negative numbers count from the end:

- `X[-1, :]` â†’ last row
- `X[:, -2:]` â†’ last two columns
- `X[:-1, :]` â†’ all rows except last

---

## Assignment and Broadcasting

You can assign values to slices:

`X[:2, :] = 11`

- selects first two rows
- fills all elements with `11`

NumPy broadcasts scalars to fit the slice automatically.

---

## Views vs Copies (Important)

Slicing usually returns a **view**, not a copy.  
Modifying a slice changes the original array.

```python
s = X[:2, :]
s[:] = 99   # modifies X too
```
